```python
pldf['Minspermatch']= pldf['Mins']/pldf['Matches'].astype(int)
pldf['GoalsperMatch']=pldf['Goals']/pldf['Matches'].astype(float)
pldf.head()

```

**Goals Scored in the Season :-**

```python
#Goals per match in the prem NOTE: this doesn't account for own goals 
print("The PL 20-21 season saw {} goals at the rate of {:.2f} goals per game".format(pldf['Goals'].sum(),(pldf['Goals'].sum())/(380)))
```

```python
#some facts on penalties
pgtotg = pldf['Penalty_Goals'].sum()/pldf['Goals'].sum()
penaltypermatch = pldf['Penalty_Attempted'].sum()/pldf['Matches'].sum()
print("{:0.2f} % of the goals scored were through penalties".format(pgtotg*100))
print("And {:0.3f} penalties were awarded per game".format(penaltypermatch)) 
```

```python
#penalties missed to penalties scored
plt.figure(figsize=(12,6))
total_pengoals = pldf['Penalty_Goals'].sum()
pen_notscored = pldf['Penalty_Attempted'].sum()- total_pengoals
data = [pen_notscored, total_pengoals]
labels = ['Penaties missed', 'Penalties scored']
color = sns.color_palette()
plt.pie(data,labels=labels,colors=color,autopct='%.2f%%',shadow=True)
plt.show()
```
